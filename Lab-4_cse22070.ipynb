```python
import pandas as pd
import numpy as np
import time
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("F:\\Amrita\\5th semester_2024\\Machine Learning\\Lab-3\\Project Dataset.xlsx")
df.head()
```

```python
df.shape
```

```python
from sklearn import preprocessing
l_c=preprocessing.LabelEncoder()
df['mentalh_level']=l_c.fit_transform(df['mentalh_level'])
```

```python
from sklearn.model_selection import train_test_split
X=df.iloc[:,1:2]
y=df['mentalh_level']
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3)
```

```python
#KNN Classifier

import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
start=time.time()
KNN=KNeighborsClassifier()
KNN.fit(X_train,y_train)
stop=time.time()
print("The accracy score of train data ",round(KNN.score(X_train,y_train),5)," and test data ",round(KNN.score(X_test,y_test),5))
print("The time taken to get the score of KNN classifier : ",round(stop-start,3))
prediction_X_test=KNN.predict(X_test)
prediction_X_train=KNN.predict(X_train)
cm = confusion_matrix(y_test,prediction_X_test)
print("The F_score of training data ",round(f1_score(y_train,prediction_X_train,average='weighted'),5))
print("The F_score of test data ",round(f1_score(y_test,prediction_X_test,average='weighted'),5))
print("The precision_score of training data ",round(f1_score(y_train,prediction_X_train,average='weighted'),5))
print("The precision_score of test data ",round(precision_score(y_test,prediction_X_test,average='weighted'),5))
print("The ROC AUC score of training data:", round(roc_auc_score(y_train, prediction_X_train,multi_class='ovo'), 5))
print("The ROC AUC score of test data:", round(roc_auc_score(y_test, prediction_X_test,multi_class='ovr'), 5))

print(cm)
```

```python
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error,r2_score
MAPE = mean_absolute_percentage_error(X_train,y_train)
MSE= mean_squared_error(X_train,y_train)
r=r2_score(X_train,y_train)
print('The Mean absolute percentage error : ',round(MAPE,3))
print('The Mean squared error : ',round(MSE,3))
print('The r2score : ',round(r,3))
```

```python
#A7
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11, 13, 15]  # Example range of k values
}
grid_search = GridSearchCV(estimator=KNN,param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train,y_train)
best_k = grid_search.best_params_['n_neighbors']
print(f"The best k value trained data : {best_k}")
```

```python
grid_search.fit(X_test,y_test)
best_k = grid_search.best_params_['n_neighbors']
print(f"The best k value test data : {best_k}")
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
param_dist = {
    'n_neighbors': randint(1, 20)  # Randomly sample from 1 to 19
}
random_search = RandomizedSearchCV(estimator=KNN, param_distributions=param_dist, n_iter=10, cv=5, scoring='accuracy', random_state=42)
random_search.fit(X_train, y_train)
best_k = random_search.best_params_['n_neighbors']
print(f"The best k value for trained data :  {best_k}")
print(f"The best score for trained data : {random_search.best_score_}")
```

```python
random_search.fit(X_test, y_test)
best_k = random_search.best_params_['n_neighbors']
print(f"The best k value for test data :  {best_k}")
print(f"The best score for test data : {random_search.best_score_}")
```
