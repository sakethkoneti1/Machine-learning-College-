```python
#A1

# Using normal probability

# Get the counts of 'buys_computer' column (number of people who bought and didn't buy a computer)
prior_1 = (df['buys_computer']).value_counts()

# Get the total number of records in the 'buys_computer' column
total_count = df['buys_computer'].count()

# Calculate the probabilities for both classes (buys and doesn't buy) by dividing counts by the total count
probability = [(prior_1[1] / total_count), (prior_1[0] / total_count)]

# Print the probabilities using normal probability calculation
print("Using normal probability: ", probability)

# Using groupby function

# Group the data by the 'buys_computer' column and calculate the relative frequency (probability) by dividing by total rows
prior = df.groupby('buys_computer').size().div(len(df))

# Print the probabilities using the groupby function
print("Using groupby function: \n", prior)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df["age"] = label_encoder.fit_transform(df["age"])
df["income"] = label_encoder.fit_transform(df["income"])
df["student"] = label_encoder.fit_transform(df["student"])
df["credit_rating"] = label_encoder.fit_transform(df["credit_rating"])
df["buys_computer"] = label_encoder.fit_transform(df["buys_computer"])
df
```

```python
#A3

# Initialize an empty list to store the matrix
matrix = []

# Initialize an empty list to store the arrays (columns from dataframe)
arr = []

# Loop through the first 4 columns (index 0 to 3) of the dataframe 'df'
for i in range(0, 4):
    # Extract the column as a numpy array and append to the 'arr' list
    arr.append(df.iloc[:, i].to_numpy())

# Outer loop to iterate through the first 4 arrays in 'arr'
for i in range(0, 4):
    # Initialize a list to store the dot products for the current row
    mat = []
    
    # Inner loop to calculate dot products for columns ahead of the current one
    for j in range(i + 1, 4):
        # Calculate the dot product between the i-th and j-th array
        mat.append(np.dot(arr[i], arr[j]))
    
    # Append the resulting row (dot products) to the matrix
    matrix.append(mat)

# Print the final matrix
print(matrix)
```

```python
F_dt=df[['age','income','student','credit_rating']]
X=np.asarray(F_dt)
y=np.asarray(df['buys_computer'])
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
#A4
from sklearn.naive_bayes import GaussianNB  # Import Gaussian Naive Bayes model from scikit-learn

model = GaussianNB()  # Initialize the Gaussian Naive Bayes model

# Calculating the accuracy for training and testing data
model = model.fit(X_train, y_train)  # Fit the model using training data (X_train and y_train)

# Testing the accuracy
T1 = model.score(X_train, y_train)  # Calculate the accuracy on the training data
T2 = model.score(X_test, y_test)  # Calculate the accuracy on the test data

# Print the accuracy results
print("The accuracy of X_train, y_train : ", T1)  # Print the accuracy of the training data
print("The accuracy of X_test, y_test : ", T2)  # Print the accuracy of the test data
```
